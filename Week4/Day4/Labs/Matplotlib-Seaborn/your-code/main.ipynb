#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100, 0.01)
y = x ** 2
z = np.array([math.exp(t) for t in x])

fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x,y)
plt.title('Not sure what I am doing', color='white', size='xx-large')
ax.set_xlabel('Am i wrong?', color='white', size='x-large')
ax.set_ylabel('Am i right?', color='white', size='x-large')

ax2 = ax.twinx()
ax2.plot(x,z, color='tab:red')

fig.tight_layout()
plt.show()

```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100, 0.01)
y = x ** 2
z = np.array([math.exp(t) for t in x])

fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x,y, color='tab:purple', label='square')
plt.title('Not sure what I am doing', color='white', size='xx-large')
ax.set_xlabel('Am i wrong?', color='white', size='x-large')
ax.set_ylabel('Am i right?', color='white', size='x-large')
ax2.legend(loc=3, fontsize='x-large')

ax2 = ax.twinx()
ax2.plot(x,z, color='tab:red', label='exp')
ax2.legend(loc=4, fontsize='x-large')

fig.tight_layout()
plt.show()

```

#### Plot the correlation matrix using seaborn.

```python
# your code here
d = titanic.copy()
# Compute the correlation matrix
corr = d.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
