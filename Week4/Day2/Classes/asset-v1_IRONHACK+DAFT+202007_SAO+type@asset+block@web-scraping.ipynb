# CS GO example

```python
n = 0

url = f'https://www.hltv.org/results?offset={n}'
url
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content)
soup
```

```python
tag_div = soup.find_all('div', attrs={'class':'result-con'})

'https://www.hltv.org' + tag_div[0].find('a')['href']
```

```python
websites = ['https://www.hltv.org' + tag_div.find('a')['href'] 
                    for tag_div in soup.find_all('div', attrs={'class':'result-con'})]
```

```python
websites[0]
```

```python
url = websites[3]
```

```python
soup = BeautifulSoup(requests.get(url).content)
```

```python
team_name1 = soup.find('div', attrs={'class':'team1-gradient'}).find('div', attrs={'class':'teamName'}).text
team_name2 = soup.find('div', attrs={'class':'team2-gradient'}).find('div', attrs={'class':'teamName'}).text
```

```python
score_team1 = soup.find('div', attrs={'class':'team1-gradient'}).find('div', attrs={'class':'lost'}).text
score_team2 = soup.find('div', attrs={'class':'team2-gradient'}).find('div', attrs={'class':'won'}).text
```

```python
score_team2
```

```python
(soup.find('div', attrs={'class':'team1-gradient'}).find('div', attrs={'class':'lost'}) or \
soup.find('div', attrs={'class':'team1-gradient'}).find('div', attrs={'class':'won'})).text
```

```python
team_name1 = soup.find('div', attrs={'class':'team1-gradient'}).find_all('div')[0].text
team_name2 = soup.find('div', attrs={'class':'team2-gradient'}).find_all('div')[0].text

team_score1 = soup.find('div', attrs={'class':'team1-gradient'}).find_all('div')[1].text
team_score2 = soup.find('div', attrs={'class':'team2-gradient'}).find_all('div')[1].text
```

```python
pd.DataFrame({'team1': [team_name1], 
              'score1': [score_team1], 
              'team2': [score_team2], 
              'score2': [team_name2]})
```

```python
url = websites[0]
soup = BeautifulSoup(requests.get(url).content)
```

```python
from tqdm.auto import tqdm
```

```python
df = pd.DataFrame()

for website in tqdm(websites):
    html = requests.get(website).content
    soup = BeautifulSoup(html)
    
    team_name1 = soup.find('div', attrs={'class':'team1-gradient'}).find_all('div')[0].text
    team_name2 = soup.find('div', attrs={'class':'team2-gradient'}).find_all('div')[0].text

    team_score1 = soup.find('div', attrs={'class':'team1-gradient'}).find_all('div')[1].text
    team_score2 = soup.find('div', attrs={'class':'team2-gradient'}).find_all('div')[1].text
        
    mini_df = pd.DataFrame({'team_a' : [team_name1],
                            'team_b' : [team_name2],
                            'team_a_score': [team_score1],
                            'team_b_score': [team_score2]})
    df = pd.concat([df, mini_df])        
```

```python
df.reset_index(drop=True)
```
